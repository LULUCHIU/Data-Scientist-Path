```python
import pandas as pd
import numpy as np
df = pd.read_excel("../@data/clinic_clean_use2.xlsx", parse_dates=["day"])
df
```

```python
df.set_index('Sample ID',inplace=True)
df
```

**1. Identify Missing Values:** Use the `isna()` method to identify missing values in your DataFrame. This will return a DataFrame of the same shape as your original DataFrame, with True for missing values and False for non-missing values.

**2. Count Missing Values:** Use the `sum()` method to count the number of missing values in each column.

```python
df.isnull().sum()
```

### There are 6 ways, you can handle your missing value

#### Method 1.
**Drop Missing Values:** Use the `dropna()` method to drop rows with any missing values.

```python
print(df.shape)
print(df.dropna().shape)
df.dropna()
```

#### Method 2.
**Fill Missing Values:** Use the `fillna()` method to fill missing values with a specified value, such as the mean or median of the column.

```python
# select numeric columns to get mean()
numeric_cols = df.select_dtypes(include=[np.number]).columns
print(numeric_cols)
```

```python
# select categorical columns to give another values or category
cate_cols = df.select_dtypes(include=['object']).columns
print(cate_cols)
```

```python
print(df[numeric_cols].fillna(value=df[numeric_cols].mean()).isnull().sum())
df[numeric_cols].fillna(value=df[numeric_cols].mean())

# df.head()
```

```python
print(df[numeric_cols].fillna(value=df[numeric_cols].median()).isnull().sum())
df[numeric_cols].fillna(value=df[numeric_cols].median())
```

- if you want to specify the value & column you chooice, you can make it with dictionary

```python
new_df = df.fillna({
    'subtype' : 'Non-viral',
    'MSIsensor Score': 0,
    'Mutation Count': df['Mutation Count'].mean()
})
new_df
```

#### Method 3
**Forward Fill or Backward Fill:** Use the `ffill()` or `bfill()` methods to fill missing values with the previous or next non-missing value.

```python
df.head()
```

```python
new_df = df.ffill()
new_df.head()
```

```python
new_df = df.bfill(limit=1) #limit: times of filled
new_df.head()
```

#### Method 4
**Interpolate:** Use the `interpolate()` method to interpolate missing values based on the values around them.

```python
# Interpolate only numeric columns
new_df = df.apply(lambda x: x.interpolate(limit_direction='both') if np.issubdtype(x.dtype, np.number) else x)
new_df
```

#### Method 5
**Replace Missing Values with Specific Values:** replace missing values with specific values.

```python
# Replace missing values in numeric columns with 0
# Replace missing values in object columns with "Unknown"
new_df = df.copy()
new_df[cate_cols] = new_df[cate_cols].fillna('Uknown')
new_df[numeric_cols] = new_df[numeric_cols].fillna(0)
new_df
```
